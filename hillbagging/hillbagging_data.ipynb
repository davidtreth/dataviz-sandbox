```python
import glob
import csv
csvfiles = glob.glob('*.csv')
print(csvfiles)
```

```python
for filename in csvfiles:
    with open(filename, newline='') as csvfile:
        spamreader = csv.reader(csvfile)
        for i, row in enumerate(list(spamreader)):
            if i == 0:
                fieldnames = row
            else:
                continue
        print(filename)
        print(fieldnames)
```

```python
hills_cornwall_name = []
hills_cornwall_name_kw = []
hills_cornwall_X = []
hills_cornwall_Y = []
hills_cornwall_Z = []

with open('hillbagging_GB_wales+cornwall.csv', newline='') as csvfile:
    spamreader = csv.DictReader(csvfile)
    for row in spamreader:
        if row['country'] != "W":
            print("{h:30s}: {k:20s}: {s:4s}: {g} :{e}, {n}: {z}m".format(h=row['hillname'], k=row['name_kw'],
                                                                   s=row['source_kw'], g=row['gridref'],
                                                                  n=row['Northing'], e=row['Easting'],
                                                                      z=row['metres']))
            hills_cornwall_name.append(row['hillname'])
            hills_cornwall_name_kw.append(row['name_kw'])
            hills_cornwall_X.append(int(row['Easting']))
            hills_cornwall_Y.append(int(row['Northing']))
            hills_cornwall_Z.append(int(row['metres']))
            
```

```python
import matplotlib.pyplot as plt
plt.plot(hills_cornwall_X, hills_cornwall_Y, "ko")
```

```python
hills_cornwall = list(zip(hills_cornwall_name, hills_cornwall_name_kw, hills_cornwall_X, hills_cornwall_Y, hills_cornwall_Z))
```

```python
hills_cornwall_gt250 = [h for h in hills_cornwall if h[4]>=250]
print(hills_cornwall_gt250)
```

```python
X = [h[2] for h in hills_cornwall_gt250]
Y = [h[3] for h in hills_cornwall_gt250]
```
